```python

plt.figure(figsize=(100, 5))
plt.xticks(fontsize=10, rotation=100)

    
    
plt.title('The situation between the Countries and their rice seeds')    
plt.bar(x, y, align='edge', width=0.5)

plt.xticks(x,a['Country'])
plt.xlabel('Countries')
plt.ylabel('The number of rice seeds')

for i, label in enumerate(list(a.index)):
    seeds = a.ix[label]['Rice']
    plt.annotate(str(seeds), (i, seeds+0))
    
plt.show()
```
